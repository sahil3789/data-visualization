###### Department wise codekata performence (pie chart)

```python
x=df.groupby(by="Department").sum("CodeKata Score")
scores = list(x["CodeKata Score"])
scores

colors = ['#FFB900','#83FF00','#FF3200']
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'CSE', 'ECE', 'EEE'

explode = (0.1, 0.1, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(scores, explode=explode, labels=labels, autopct='%1.1f%%',colors=colors,
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
